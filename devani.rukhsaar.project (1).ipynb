## Python Web Scraping Project 
### Algorithm

1. Please restart and clear the kernel
2. For the User input, please only use numbers from 1-10 for the program to work, or the program will break and exit
3. Import BeautifulSoup to extract data from HTML files, in our case the apple.com's top chart URLs which have been saved in a URL list. 
4. Algorithm creates a file called Ranking.txt on user's computer so please make sure Jupyter has permissions to read and write on your computer. 
5. The Rankings.txt is a comma separated file
6. Any input less than or greater than 10 will exit the program.
   Any input other than integers will end the program
7. Any input for the ranking less than 1 or greater than 100 will give a python error. The rankings are for top 100s so please keep your inputs within that range.
8. To Exit the program type 10 in the user input section.


```python
import requests              #importing desired programs
from bs4 import BeautifulSoup
import pandas as pd
import os, subprocess
import glob
chart_type_list=["songs","albums","free-apps","paid-apps","top-grossing-apps","paid-books","movies","tv-shows","music-videos"]
url="https://www.apple.com/itunes/charts/{chart_type}/"#creating a list of all the pages we want
urls=list()  #creating a new list to append URLs


list2=[]

for my_chart_type in chart_type_list:   #appending the pages in URL using for loop
    urls.append(url.format(chart_type=my_chart_type))
    
for x in urls:
    y= requests.get(x)
    soup=BeautifulSoup(y.text,"html.parser")
    results=soup.find_all("li")
   
   
    for result in results[17:117]:  #webscrapping using for loop. the ranking is b/w strong, name in heading 3 and author/singer in heading 4
        number=result.find("strong").text[0:-1]
        name=result.find("h3").text[0:]
        author=result.find("h4").text[0:] 
        list2.append([number,name,author]) #appending the number name and author in list 2
fname = open("Rankings.txt",'w',encoding="utf-8") #writing a new file name ratings
list1 = list2[0:100]
list3 = list2[100:200]
list4 = list2[200: 300]
list5 = list2[300:400]
list6 = list2[400:500]
list7 = list2[500:600]
list8 = list2[600:700]
list9 = list2[700:800]
list10 = list2[800:900]
fname.write("Songs\n")   #writing in the files  
for i in range(0,100):
        fname.write(list1[i][0])
        fname.write(",")
        fname.writelines(list1[i][1]+","+list1[i][2])
        fname.write("\n")
    
fname.write("Albums\n")
for i in range(0,100):
        fname.write(list3[i][0])
        fname.write(",")
        fname.writelines(list3[i][1]+","+list3[i][2])
        fname.write("\n")

fname.write("Free apps\n")
for i in range(0,100):
        fname.write(list4[i][0])
        fname.write(",")
        fname.writelines(list4[i][1]+","+list4[i][2])
        fname.write("\n")
fname.write("Paid apps\n")
for i in range(0,100):
        fname.write(list5[i][0])
        fname.write(",")
        fname.writelines(list5[i][1]+","+list5[i][2])
        fname.write("\n")
fname.write("Top Grossing Apps\n")
for i in range(0,100):
        fname.write(list6[i][0])
        fname.write(",")
        fname.writelines(list6[i][1]+","+list6[i][2])
        fname.write("\n")
        
fname.write("Books\n")
for i in range(0,100):
        fname.write(list7[i][0])
        fname.write(",")
        fname.writelines(list7[i][1]+","+list7[i][2])
        fname.write("\n")

fname.write("movies\n")
for i in range(0,100):
        fname.write(list8[i][0])
        fname.write(",")
        fname.writelines(list8[i][1]+","+list8[i][2])
        fname.write("\n")    
        
        
fname.write("TV shows\n")
for i in range(0,100):
        fname.write(list9[i][0])
        fname.write(",")
        fname.writelines(list9[i][1]+","+list9[i][2])
        fname.write("\n") 
        
fname.write("Music videos\n")
for i in range(0,100):
        fname.write(list10[i][0])
        fname.write(",")
        fname.writelines(list10[i][1]+","+list10[i][2])
        fname.write("\n") 
fname.close()


        
print("1. Songs\n2. Albums\n3. Free Applications\n4. Paid Applications \n5.Top Grossing Applications\n6. Books\n7. Movies\n8. TV Shows\n9. Music Videos\n10. Exit")
while True: #using while loop to keep the program running till the user doesnt ask it
    user_input4=input("Please enter any number between 1 to 10 depending on the category you want to choose ")
    try: #using try and error if the input is something other than the number
        user_input=int(user_input4)
        user_input>0 and user_input<11
    
        if user_input>0 and user_input<=9: #printing the choice of the user
            print(chart_type_list[user_input-1])
            user_input3=input("Please enter the rating you want between 1 and 100 ")
            try:
                user_input1=int(user_input3)#asking for ranking 
                if(user_input==1):
                    print("information about ranking",user_input1, "is\n  Ranking, name,  singer\n", list2[user_input1-1])
                elif(user_input>=2 or user_input<=9):
                    print("information about ranking",user_input1, "is\n  Ranking, name,  author/TVshow name\n",list2[(user_input-1)*100+user_input1-1])
                else:
                    print("Good bye")
                    exit
#using break to break the condition and get out of the loop         
            except:
                print("Please enter a number between 1 and 100 and try again. GOOD BYE")
                exit
                break
        else:
            print("Good bye")
            exit
            break
            
    except:
        print("Please enter a number between 1 and 10 and try again. GOOD BYE")
        exit
        break
```
